```python
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
from PIL import Image
import requests
import torch

# Load image
url = 'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcR-0zYMMI8h_pwan2Cu8SzcR5Xo2FmWgaO5LA&s'
image = Image.open(requests.get(url, stream=True, timeout=30).raw).convert("RGB")

# ðŸ‘‡ Load from local directory instead of downloading
model_path = "C:/Users/baher/OneDrive/Desktop/masters/models/trocr-small-stage1"

processor = TrOCRProcessor.from_pretrained(model_path)
model = VisionEncoderDecoderModel.from_pretrained(model_path)

# Inference
pixel_values = processor(image, return_tensors="pt").pixel_values
decoder_input_ids = torch.tensor([[model.config.decoder.decoder_start_token_id]])
outputs = model(pixel_values=pixel_values, decoder_input_ids=decoder_input_ids)
```
